```python
import requests, zipfile
import os
from pathlib import Path

num_of_downloads = 35
for i in range(num_of_downloads):
    r = requests.get(dataset_urls[i], stream=True)
    if not os.path.isfile(f'zipped/{i}.zip'):
        Path('zipped').mkdir(parents=True, exist_ok=True)
        with open(f'zipped/{i}.zip', 'wb') as fd:
            for chunk in r.iter_content(chunk_size=128):
                fd.write(chunk)
```

Convert images to video

```python
from glob import glob
from pathlib import Path
import cv2
import re
import imageio

def images_to_video(images_path: str, video_path_save: str, fps = 240):
    images = []
    paths = list(glob(str(Path(images_path) / "*.*")))
    for path in sorted(paths):
        images.append(cv2.imread(path))

    print(len(images))

    with imageio.get_writer(video_path_save, mode='I', fps=fps, codec='libx264') as writer:
        for im in images:
            writer.append_data(im[..., ::-1])


images_to_video('zipped/bee/30/bee', 'videos/bee_1.mp4', fps=30)   
```
